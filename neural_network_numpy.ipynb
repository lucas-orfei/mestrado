```python
# Separação dos dados de teste e treino

from sklearn.model_selection import train_test_split

# Separação dos dados mas mantendo a dimensão
def custom_train_test_split(X, y, test_size=0.2):
    X_train, X_test, y_train, y_test = [], [], [], []

    for i in range(len(X)):
        # Realiza o split para cada lista interna
        X_tr, X_te, y_tr, y_te = train_test_split(X[i], y[i], test_size=test_size)
        
        # Adiciona os resultados nas listas
        X_train.append(X_tr)
        X_test.append(X_te)
        y_train.append(y_tr)
        y_test.append(y_te)
    
    return X_train, X_test, y_train, y_test

# Realiza a separação
X_train, X_test, Y_train, Y_test = custom_train_test_split(density, composition, test_size=0.2)

# Converte as listas para numpy arrays
X_train = np.array(X_train).T
X_test = np.array(X_test).T
Y_test = np.array(Y_test).T
Y_train = np.array(Y_train,dtype = int).T

Y_train_flat = np.array([item for sublist in Y_train for item in sublist])
num_classes = 3  

# Converte para one-hot encoding
Y_train_one_hot = np.eye(num_classes)[Y_train_flat]
Y_train_one_hot = Y_train_one_hot.reshape(100, 16, num_classes)
#Y_train_one_hot = Y_train_one_hot.reshape(-1, num_classes)
```

```python
def one_hot_encode_and_flatten(y, num_classes=3):
    # Cria um array de zeros com forma (samples, tanques, classes)
    y_one_hot = np.eye(num_classes)[y]
    # Achata para (samples, tanques * classes)
    return y_one_hot.reshape(y.shape[0], -1)

# Converter para one-hot e achatar
Y_train_encoded = one_hot_encode_and_flatten(Y_train)

# Y_train_encoded agora tem a forma (16, 300)
print(Y_train_encoded.shape)
```

Vamos usar a arquitetura Feed-Forward Network pois teremos camadas sucessivas que alimentam as próximas - começando na camada de entrada e terminando na camada de saída. Todos os nós da rede estão conectados aos da próxima camada. 

Para um vetor *z* de dimensão *K* classes, temos que a função softmax é definida como
$$softmax(z_i) = \frac{exp(z_i)}{\sum_{j = 1}^{K} exp(z_j)}, 1 \leq i \leq K$$ 

Ainda, temos que o vetor *z* é definido por

$$z = \left(\sum_{i = 1}^{n}w_ix_i\right) + b $$

onde, $x_i$ são as entradas, $w_i$ são os pesos e *b* é o bias (viés). Sendo assim, a função softmax transforma o vetor *z* em probabilidades entre 0 e 1.

Para treinar o modelo, vamos utilizar a **Função de perda de Entropia Cruzada (Cross-Entropy Loss)**. A distância entre a saída do modelo e a saída real é chamada de função de perda ou de custo. E portanto, temos que minimizar essa distância, ou seja, essa perda, ao máximo para que o modelo tenha resultados melhores. Na forma matemática, estaremos calculando os pesos $w_i$ e o bias *b* com a função de perda e depois otimizando esses valores minimizandos-os com o **Gradiente Descendente Estocástico.**

```python
# Arquitetura que será usada: feed-forward network
# Qual será o tamanho da rede neural? Digo, tamanho de cada camada. 
# Quais pesos vou inicializar?
# Qual função de ativação vou usar? Deve ser não linear Softmax
# Como minimizar os pesos? Função de Perda deve ser diferenciável, isso implica que a função de ativação também deve ser diferenciável.
# Gradiente descendente
# Backpropagation
```

```python
class NeuralNetwork:
    def __init__(self, n_inputs, n_neurons):
    # Inicialização dos pesos e bias
        self.weights = 0.01 * np.random.randn(n_inputs, n_neurons)
        self.biases = np.zeros((1, n_neurons))

    def forward(self, inputs):
    # Calcula os valores de saída de cada camada
        self.inputs = inputs
        self.output = np.dot(inputs, self.weights) + self.biases # X.W + b

    def backward(self, dvalues):
        # Gradiente dos parâmetros
        self.dweights = np.dot(self.inputs.T, dvalues)
        self.dbiases = np.sum(dvalues, axis = 0, keepdims = True)
        # Gradiente dos valores
        self.dinputs = np.dot(dvalues, self.weights.T)
```

```python
class ReLU():
    def forward(self, inputs):
        self.inputs = inputs
        self.output = np.maximum(0, inputs) # Função ReLU R(x) = max(0,x)
    def backward(self,dvalues):
        self.dinputs = dvalues.copy()
        self.dinputs[self.inputs <= 0] = 0
```

```python
class Softmax():
    def forward(self,inputs):
        # Probabilidades não normalizadas
        exp_values = np.exp(inputs - np.max(inputs, axis = 1, keepdims = True))
        # Normalização
        probabilities = exp_values / np.sum(exp_values, axis = 1, keepdims = True)
        self.output = probabilities
```

```python
class Loss:
    def calculate(self, output, y):
        sample_losses = self.forward(output, y)
        data_loss = np.mean(sample_losses)
        return data_loss
```

```python
class LossCategoricalCrossentropy(Loss):
    # Forward pass
    def forward(self, y_pred, y_true):
        samples = len(y_pred)
        y_pred_clipped = np.clip(y_pred, 1e-7, 1 - 1e-7)

        if len(y_true.shape) == 1:
            correct_confidences = y_pred_clipped[
                range(samples),
                y_true
            ]

        elif len(y_true.shape) == 2:
            correct_confidences = np.sum(
                y_pred_clipped * y_true,
                axis=1
            )
        else:
            raise ValueError("Formato de y_true não suportado.")
        
        negative_log_likelihoods = -np.log(correct_confidences)
        return negative_log_likelihoods


    def backward(self, dvalues, y_true):

        samples = len(dvalues)
        labels = len(dvalues[0])

        if len(y_true.shape) == 1:
            y_true = np.eye(labels)[y_true]

        # Gradiente
        self.dinputs = -y_true / dvalues
        # Normalização
        self.dinputs = self.dinputs / samples
```

```python
class Activation_Softmax_Loss_CategoricalCrossentropy:
    
    def __init__(self):
        self.activation = Softmax()
        self.loss = LossCategoricalCrossentropy()

    def forward(self, inputs, y_true):
        self.activation.forward(inputs)
        self.output = self.activation.output
        return self.loss.calculate(self.output, y_true)

    def backward(self, dvalues, y_true):
        samples = len(dvalues)

        if len(y_true.shape) == 2:
            y_true = np.argmax(y_true, axis=1)

        self.dinputs = dvalues.copy()
        # Gradiente
        self.dinputs[range(samples), y_true] -= 1
        # Normalização
        self.dinputs = self.dinputs / samples
```

```python
# Gradiente Descendente 
class Optimizer_GD:
 
 def __init__(self, learning_rate=0.01):
  self.learning_rate = learning_rate

 def update_params(self, layer):
  layer.weights += -self.learning_rate * layer.dweights
  layer.biases += -self.learning_rate * layer.dbiases
```

```python
# Cria a primeira camada
layer1 = NeuralNetwork(100, 64)

# Cria a função de ativação ReLU
activation1 = ReLU()

# Cria uma segunda camada
layer2 = NeuralNetwork(64, 300)

# Cria a função perda e softmax
loss_activation = Activation_Softmax_Loss_CategoricalCrossentropy()

# Cria o gradiente descendente
optimizer = Optimizer_GD()

# Treinamento em loop
for epoch in range(10000):
    # Com os dados de treino, aplicando o método forward
    layer1.forward(X_train)
    
    # Usa a função de ativação da saída da primeira camada
    activation1.forward(layer1.output)
    
    # Passe os resultados para a próxima camada
    layer2.forward(activation1.output)
    
    # Usa os dados da segunda camada para a função de perda e softmax
    loss = loss_activation.forward(layer2.output, Y_train_encoded)
    
    # Calcula a precisão
    predictions = np.argmax(loss_activation.output, axis=1)
    if len(Y_train_encoded.shape) == 2:
        Y_train_encoded = np.argmax(Y_train_encoded, axis=1)
    accuracy = np.mean(predictions == Y_train_encoded)
    
    if not epoch % 100:
        print(f'epoch: {epoch}, ' +
              f'acc: {accuracy:.3f}, ' +
              f'loss: {loss:.3f}')
    
    # Backward
    loss_activation.backward(loss_activation.output, Y_train_encoded)
    layer2.backward(loss_activation.dinputs)
    activation1.backward(layer2.dinputs)
    layer1.backward(activation1.dinputs)
    
    # Atualiza o peso e bias para as camadas 1 e 2
    optimizer.update_params(layer1)
    optimizer.update_params(layer2)
```
